```python
import torch
import torchaudio
import IPython.display as ipd
from include.data_loader import read_json
```

```python
command = torch.from_numpy(read_json("./data/Fylhtq/on.json"))
base_sample_rate = command.shape[-1]
new_sample_rate = 8000
transform = torchaudio.transforms.Resample(orig_freq=base_sample_rate, new_freq=new_sample_rate)
```

```python
command.shape
```

```python
ipd.Audio(command.numpy(), rate=5000)
```

```python
command_t = transform(command)
ipd.Audio(command_t.numpy(), rate=new_sample_rate)
```
